#### 9. How many Counties have 2010 populations greater than 1 million?

```python
# Write your code here
def population_bracket(x):
    if x > 1000000:
        return "More than 1 Million"
    else:
        return "Less than 1 Million"
```

```python
total_countries=top5_countries.groupby(['County']).sum()
total_countries['population_bracket']=total_countries['2010Census'].apply(population_bracket)
total_countries_1M=total_countries[total_countries['population_bracket']!='Less than 1 Million']
total_countries_1M['2010Census'].nunique()
```

#### 12. What States have the highest estimated percent change between the 2010 Census and the 2017 Population Estimate?

**NOTE:** This will take several lines of code, as it requires a recalculation of Percent Change.
